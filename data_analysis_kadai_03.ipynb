```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

# read csv file
df = pd.read_csv("sample_pandas_6.csv")
category_df =pd.read_csv('category.csv')

# merge two dataframe
df =pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

# set variable of bar graph
x = [1,2,3,4,5,6]
y_s = df['カテゴリー'].value_counts()
# type(df['カテゴリー'].value_counts())
#plt.bar(x, y)
y = [y_s[0], y_s[1], y_s[2], y_s[3], y_s[4], y_s[5]]

# write bar graph
label = y_s.index
label
plt.title("Quantity of each category")
plt.bar(x, y, tick_label=label)
plt.show()

# description of orders of each product number
print('\n\n')
df.groupby('商品番号')['注文数'].describe()
```
