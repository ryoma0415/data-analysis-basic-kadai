```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2, 3, tight_layout=True)

y0 = [1, 2, -5, 2]
ax[0,0].plot(y0)

x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
ax[0,1].plot(y1)

y2 = np.random.normal(15, 5, 2000)
ax[0,2].hist(y2)

np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))
ax[1,0].scatter(x3, y3)

plt.show()
```
